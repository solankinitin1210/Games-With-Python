**What is Secret Auction Program ?**

It is also known as blind auction. In this type of auction, all bidders simultaneously submit sealed bids so that no bidder knows the bid of any other participant. The highest bidder pays the price that was submitted.

In This Program We'll Automate This Procress Thru Python Code

auction.jpg

**Process We'll Follow In This Program : **

*   We'll Take Use Name and His/Her Bid Amount
*   Store User Name And Bid to Dictionary
*   We'll Ask User If Ther is Any Other Bidder if yes Then We'll Clear Previus Data And Take Another Bidder Data
*   We'll Continue Untill all Bidders Enter Their Bids
*   Finaly We'll Display The Name of Winner And His/Her Bid Amount.





```python
from IPython.display import clear_output
logo = '''
                         ___________
                         \         /
                          )_______(
                          |"""""""|_.-._,.---------.,_.-._
                          |       | | |               | | ''-.
                          |       |_| |_             _| |_..-'
                          |_______| '-' `'---------'` '-'
                          )"""""""(
                         /_________\\
                       .-------------.
                      /_______________\\
'''
bid_dict = {}
more_input = 'y'
while more_input == 'y':
  print(logo)
  name = input('Enter Name Of Person : ')
  no_error = True
  while no_error :
    try :
      bid = int(input('Enter Bid from that Person : '))
      no_error = False
    except :
      print('Wrong Input!! Try Again!')
  bid_dict[bid] = name
  more_input = input('Is there any other who wants to bid? yes or no : ').lower()
  clear_output()

sum = 0
for i in bid_dict :
  if i > sum :
    sum = i
print(f'{bid_dict[sum]} won the bid with bid amount {sum}.')

```
