## Problem statement ##
Create famous 'Frog leap' puzzle game. Try completing the game before starting to get an idea about its working.
[Demonstration](https://www.neok12.com/games/leap-froggies/leap-froggies.htm).<br>

### Rules ###
1. The left set of frogs can only move right, the right set of frogs can only move left.
2. Frogs can move forward one space, or move two spaces by jumping over another frog from opposite side.
3. The puzzle is solved when the two sets of frogs have switched positions.
<br>



## Steps to solve the problem:
### Step1:-
- Display green and brown frogs on the left and right sides initially.

Initial Display :-  
```
[ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]
['G', 'G', 'G', '-', 'B', 'B', 'B']
```
<br>
Here 'G' represents Green frogs on the left side and 'B' represents brown frogs on the right side. The '-' defines the position of empty leaf.
(You can change display according to your imagination or convinience)

### Step2:-
Accept positions of the frog that you want to move.<br>
Example: If we enter position 2 then the game will look like this:-
```
[ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]
['G', 'G', '-', 'G', 'B', 'B', 'B']
```

### Step3:- ###
Define Invalid moves and add conditional 'if' statements accordingly
#### Rules
1. Entered position should be between 0 to 6. It can be a character 'q' to quit the game.
2. Entered position cannot be the position of empty leaf.
3. If the selected frog position cannot perform the contrains given in rule 2 then the move is invalid.

### Step4:-
Make the appropriate move by changing the game display.

````python
### Your code ###
# Create class for colors for tesx
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

def start():
    # Define all the variables and emojis which will use in program
    global g,b,s,d,d1,input_list,kl,counter
    g  = '\U0001F438'
    b  = '\U0001F42C'
    s  = '\U0001F341'
    # Set Initial Position of frogs
    input_list=['1','2','3','4','5','6','7','q','Q','r','R']
    d={1:g,2:g,3:g,4:s,5:b,6:b,7:b}
    d1={1:b,2:b,3:b,4:s,5:g,6:g,7:g}
    kl=list(d.keys())
    counter=0

    # Print Games Rule
    print(color.BOLD+ color.RED + 'Rules Of Game' + color.END + color.END)
    print('''    1. The left set of frogs can only move right, the right set of frogs can only move left.
    2. Frogs can move forward one space, or move two spaces by jumping over another frog from opposite side.
    3. The puzzle is solved when the two sets of frogs have switched positions.
    4. Entered position should be between 0 to 6. It can be a character 'q' to quit the game.
    5. Entered position cannot be the position of empty leaf.
    6. If the selected frog position cannot perform the contrains given in rule 2 then the move is invalid.''')

    # Print All Commands
    print(color.BOLD+ '\nUse Below Commands to play game'+ color.END)
    print('To Move Frog :',color.BOLD+ color.BLUE + '1,2,3,4,5,6,7' + color.END + color.END, ' | Restart :',color.BOLD+ color.BLUE + 'r' + color.END + color.END, ' | Quit :',color.BOLD+ color.BLUE + 'q' + color.END + color.END)

    # Print Initial Position of missionaries,cannibas and boat
    print(color.BOLD + '\nLets Start the Game' + color.END)
    print_status()


def print_status():
 print('No. of move :',counter)
 print('',kl[0],'  ',kl[1],'   ',kl[2],'  ',kl[3],'   ',kl[4],'  ',kl[5],'   ',kl[6])
 print(d[1],' ',d[2],' ',d[3],' ',d[4],' ',d[5],' ',d[6],' ',d[7])

def invalid_move():
 w='\U0001F6AB'
 print('',w*3,color.BOLD + 'Invalid Move ! Please try other move' + color.END,w*3)

def won():
 print_status()
 n='\U0001F389'
 print(' \n',n*11,color.BOLD + color.GREEN + '\n Hurreyyy !! YOU WON THE GAME \n' + color.END,n*11)

def lost():
 n='\U0001F389'
 print(' \n',n*11,color.BOLD + color.GREEN + '\n Hurreyyy !! YOU WON THE GAME \n' + color.END,n*11)

def quit():
 e='\U0001F44B'
 print(e*5,color.BOLD + color.DARKCYAN + 'BYE BYE' + color.END,e*5)

start()
while True :
    x=input('\nYour Move : ')
    counter=counter+1

    if x.upper()=='Q':
     quit()
     break

    elif x.upper()=='R':
     start()
     continue

    elif x not in input_list:
     invalid_move();counter=counter-1
     continue

    x=int(x)

    if (d.get(x)== b and d.get(x-1)==s) or (d.get(x)== g and d.get(x+1)==s)  :
     if d.get(x-1)==s:
      d[x],d[x-1]=d[x-1],d[x]
     elif d.get(x+1)==s :
      d[x],d[x+1]=d[x+1],d[x]

    elif ((d.get(x)== b and d.get(x-2)==s) or (d.get(x)== g and d.get(x+2)==s)):
     if d.get(x-2)==s:
      d[x],d[x-2]=d[x-2],d[x]
     elif d[x+2]==s :
      d[x],d[x+2]=d[x+2],d[x]

    else :
     invalid_move();counter=counter-1
     continue

    if d==d1:
     won()
     break

    print_status()

# 1-2 3-3 3-2 1
# 3 5 6 4 2 1 3 5 7 6 4 2 3 5 4
````
